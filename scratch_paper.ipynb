```python
import pandas as pd
```

```python
file = "data/database.csv"
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
hss = df[df["Type"] == "HSS"]
hss.head()
```

```python
df.loc[(df["EDI_Std_Nomenclature"].str.count("X") > 1) & (df["Type"] == "HSS"), ["Type"]] = "RECT HSS"
```

```python
hss = df[df["Type"] == "RECT HSS"]
hss.head()
```

```python
df.loc[(df["EDI_Std_Nomenclature"].str.count("X") == 1) & (df["Type"] == "HSS"), ["Type"]] = "ROUND HSS"
```

```python
hss = df[df["Type"] == "ROUND HSS"]
hss.head()
```

```python
df.replace(to_replace="NAN", value="NaN", inplace=True)
df.head()
```
